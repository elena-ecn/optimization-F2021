### Problem 1 (10 Points)

Sketch graphically the problem 

$$
\begin{aligned}
\min_{x_1,x_2} & \quad f({\bf x})=(x_1+1)^2+(x_2-2)^2\\
{\text{subject to }} & \quad g_1 = x_1-2\leq 0,{\quad} g_3 = -x_1\leq 0,\\
& \quad g_2 = x_2-1\leq 0, {\quad} g_4 = -x_2\leq 0.
\end{aligned}
$$

Find the optimum graphically. Determine directions of feasible descent at the corner points of the feasible domain. Show the gradient directions of $f$ and $g_i$s at these points. Verify graphical results analytically using the KKT conditions.

---
#### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle



def f(x):
    """The objective function."""
    return (x[0]+1)**2 + (x[1]-2)**2

def plot_contour(f):
    """Plots objective function contour, constraints and gradient directions."""

    # Create mesh
    x1 = np.linspace(-4,4,100)
    x2 = np.linspace(-4,4,100)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f([X1, X2])

    # Contour
    fig, ax = plt.subplots(figsize = (12,8))
    plt.contourf(X1,X2,Z,10, cmap='OrRd')
    ax.set(xlabel='x1', ylabel='x2')
    ax.set_title('Contour of f(x), Constraints and Gradient Directions')
    plt.colorbar()
    # Axes
    ax.axhline(y=0, linewidth=1, color='k')  
    ax.axvline(x=0, linewidth=1, color='k')
    # Constraints
    ax.hlines(y=1, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.hlines(y=0, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.vlines(x=0, ymin=0, ymax=1, linewidth=1.5, color='k')
    ax.vlines(x=2, ymin=0, ymax=1, linewidth=1.5, color='k')
    # Optimum
    ax.plot(0, 1, 'o', color='r', label='Constrained Optimum')
    ax.plot(-1, 2, 'o', color='#ffb3b0', label='Obj. Funtion Optimum')
    # Gradient directions
    c = 5  # Factor to reduce the vector magnitudes
    plt.arrow(0,1, 2/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,1)
    plt.arrow(0,1, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 2/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,0)
    plt.arrow(0,0, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 6/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,0)
    plt.arrow(2,0, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 6/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,1)
    plt.arrow(2,1, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.legend()
    # Anotations and feasible domain fill
    ax.annotate(r'$\frac{\partial f}{\partial x}$', xy=(2.8, 1),color='#ab4343', size='x-large')
    ax.annotate('feasible domain', xy=(0.2, 0.2),color='#702c2c', size='large')
    ax.add_patch(Rectangle((0, 0), 2, 1, color="#702c2c", alpha=0.2))
    plt.show()


plot_contour(f)
```

#### KKT Conditions

The Lagrangian is defined as:

$$\begin{aligned}
    \mathcal{L}=(x_1+1)^2+(x_2-2)^2+\mu_1 (x_1-2) +\mu_2 (x_2-1) +\mu_3 (-x_1) +\mu_4 (-x_2)
  \end{aligned}
$$

The gradient of the Lagrangian is:


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}
    \end{bmatrix}=
    \begin{bmatrix}
    2(x_1+1) + \mu_1 - \mu_3\\\\
    2(x_2-2) + \mu_2 - \mu_4
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    0
    \end{bmatrix}
$$


Since there are 4 constraints, there are $2^4$ possibilities for each $\mu_i >0$ or $\mu_i =0$. However, since we have solved the problem graphically, we will only check this point to see whether it satisfies the KKT conditions.

At the point $(x_1,x_2)=(0,1)$, we have the following active constraints: $g_2$ and $g_3$. Therefore, 

$$\begin{aligned}
    \mu_2 >0, \mu_3 >0, \mu_1 =\mu_4 =0
  \end{aligned}
$$


<br>
$$\Rightarrow 
  \begin{equation*}
  \left\{\begin{aligned}
    2(x_1+1) - \mu_3 =0\\
    2(x_2-2) + \mu_2 =0\\
    -x_1 = 0\\
    x_2 -1 = 0
  \end{aligned}
  \right.\end{equation*} 
$$


<br>
$$\Rightarrow 
  \begin{equation*}
    \mu_2 = 2, \mu_3 = 2 >0
  \end{equation*} 
$$

This point satisfies the KKT conditions, hence it is a local solution.

The feasible domain (rectangle) and the objective function (circular contours) are convex, thus the problem is convex. Therefore, since the problem is convex, the local solution $(x_1,x_2)=(0,1)$ is the global optimum.
