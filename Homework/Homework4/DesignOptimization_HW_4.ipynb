### Problem 1 (10 Points)

Sketch graphically the problem 

$$
\begin{aligned}
\min_{x_1,x_2} & \quad f({\bf x})=(x_1+1)^2+(x_2-2)^2\\
{\text{subject to }} & \quad g_1 = x_1-2\leq 0,{\quad} g_3 = -x_1\leq 0,\\
& \quad g_2 = x_2-1\leq 0, {\quad} g_4 = -x_2\leq 0.
\end{aligned}
$$

Find the optimum graphically. Determine directions of feasible descent at the corner points of the feasible domain. Show the gradient directions of $f$ and $g_i$s at these points. Verify graphical results analytically using the KKT conditions.

---
#### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle



def f(x):
    """The objective function."""
    return (x[0]+1)**2 + (x[1]-2)**2

def plot_contour(f):
    """Plots objective function contour, constraints and gradient directions."""

    # Create mesh
    x1 = np.linspace(-4,4,100)
    x2 = np.linspace(-4,4,100)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f([X1, X2])

    # Contour
    fig, ax = plt.subplots(figsize = (12,8))
    plt.contourf(X1,X2,Z,10, cmap='OrRd')
    ax.set(xlabel='x1', ylabel='x2')
    ax.set_title('Contour of f(x), Constraints and Gradient Directions')
    plt.colorbar()
    # Axes
    ax.axhline(y=0, linewidth=1, color='k')  
    ax.axvline(x=0, linewidth=1, color='k')
    # Constraints
    ax.hlines(y=1, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.hlines(y=0, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.vlines(x=0, ymin=0, ymax=1, linewidth=1.5, color='k')
    ax.vlines(x=2, ymin=0, ymax=1, linewidth=1.5, color='k')
    # Optimum
    ax.plot(0, 1, 'o', color='r', label='Constrained Optimum')
    ax.plot(-1, 2, 'o', color='#ffb3b0', label='Obj. Funtion Optimum')
    # Gradient directions
    c = 5  # Factor to reduce the vector magnitudes
    plt.arrow(0,1, 2/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,1)
    plt.arrow(0,1, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 2/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,0)
    plt.arrow(0,0, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 6/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,0)
    plt.arrow(2,0, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 6/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,1)
    plt.arrow(2,1, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.legend()
    # Anotations and feasible domain fill
    ax.annotate(r'$\frac{\partial f}{\partial x}$', xy=(2.8, 1),color='#ab4343', size='x-large')
    ax.annotate('feasible domain', xy=(0.2, 0.2),color='#702c2c', size='large')
    ax.add_patch(Rectangle((0, 0), 2, 1, color="#702c2c", alpha=0.2))
    plt.show()


plot_contour(f)
```

#### KKT Conditions

The Lagrangian is defined as:

$$\begin{aligned}
    \mathcal{L}=(x_1+1)^2+(x_2-2)^2+\mu_1 (x_1-2) +\mu_2 (x_2-1) +\mu_3 (-x_1) +\mu_4 (-x_2)
  \end{aligned}
$$

The gradient of the Lagrangian is:


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}
    \end{bmatrix}=
    \begin{bmatrix}
    2(x_1+1) + \mu_1 - \mu_3\\\\
    2(x_2-2) + \mu_2 - \mu_4
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    0
    \end{bmatrix}
$$


Since there are 4 constraints, there are $2^4$ possibilities for each $\mu_i >0$ or $\mu_i =0$. However, since we have solved the problem graphically, we will only check this point to see whether it satisfies the KKT conditions.

At the point $(x_1,x_2)=(0,1)$, we have the following active constraints: $g_2$ and $g_3$. Therefore, 

$$\begin{aligned}
    \mu_2 >0, \mu_3 >0, \mu_1 =\mu_4 =0
  \end{aligned}
$$


<br>
$$\Rightarrow 
  \begin{equation*}
  \left\{\begin{aligned}
    2(x_1+1) - \mu_3 =0\\
    2(x_2-2) + \mu_2 =0\\
    -x_1 = 0\\
    x_2 -1 = 0
  \end{aligned}
  \right.\end{equation*} 
$$


<br>
$$\Rightarrow 
  \begin{equation*}
    \mu_2 = 2, \mu_3 = 2 >0
  \end{equation*} 
$$

This point satisfies the KKT conditions, hence it is a local solution.

The feasible domain (rectangle) and the objective function (circular contours) are convex, thus the problem is convex. Therefore, since the problem is convex, the local solution $(x_1,x_2)=(0,1)$ is the global optimum.


### Problem 2 (10 Points)

Graph the problem 

$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-x_1\\
{\text{subject to }} & \quad g_1=x_2-(1-x_1)^3\leq 0{\quad} {\rm and}{\quad} x_2\geq 0.
\end{aligned}
$$ 

Find the solution graphically. Then apply the optimality conditions. Can you find a solution based on the optimality conditions? Why? (From Kuhn and Tucker, 1951.)

---
#### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle



def f(x):
    """The objective function."""
    return -x[0]

def plot_contour(f):
    """Plots objective function contour, constraints and gradient directions."""

    # Create mesh
    x1 = np.linspace(-4,4,100)
    x2 = np.linspace(-4,4,100)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f([X1, X2])

    # Contour
    fig, ax = plt.subplots(figsize = (12,8))
    plt.contourf(X1,X2,Z,10, cmap='OrRd')
    ax.set(xlabel='x1', ylabel='x2')
    ax.set_title('Contour of f(x) and Constraints')
    plt.colorbar()
    # Axes
    ax.axhline(y=0, linewidth=1, color='k')  
    ax.axvline(x=0, linewidth=1, color='k')
    # Constraints
    y = (1-x1)**3
    ax.plot(x1, y, color='k')
    ax.hlines(y=0, xmin=-4, xmax=4, linewidth=2, color='k')
    # Optimum
    ax.plot(1, 0, 'o', color='#702c2c', label='Constrained Optimum')    
    # Anotations and feasible domain fill
    ax.annotate('feasible domain', xy=(-3, 1),color='k', size='large')
    ax.fill_between(x1, y, where=x1<1, color='#539ecd', alpha=0.3)
    ax.set_ylim([-4, 4])
    plt.legend()
    plt.show()


plot_contour(f)
```

From the graph above, we observe that the solution is at $(x_1,x_2)=(1,0)$.

The Lagrangian is defined as:

$$\begin{aligned}
    \mathcal{L}= -x_1+\mu_1 (x_2-(1-x_1)^3) +\mu_2 (-x_2) 
  \end{aligned}
$$

The gradient of the Lagrangian is:


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}
    \end{bmatrix}=
    \begin{bmatrix}
    -1+3 \mu_1 (1-x_1)^2\\\\
    \mu_1 - \mu_2
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    0
    \end{bmatrix}
$$


At $(x_1,x_2)=(1,0)$, which we know is the solution, both constraints are active, thus:

$$\begin{aligned}
    \mu_1 >0, \mu_2 >0
  \end{aligned}
$$


Therefore, from the gradient of the Lagrangian we get that:

<br>
$$\Rightarrow 
  \begin{equation*}
  \left\{\begin{aligned}
    -1  =0\\
    \mu_1 =\mu_2
  \end{aligned}
  \right.\end{equation*} 
$$

This does not hold, hence this point does not satisfy the KKT conditions. We observe that an optimal solution does exist but the KKT conditions are not satisfied.

The constraint gradients are computed as follows:

$$\frac{\partial g_1}{\partial x}=
    \begin{bmatrix}
    3(1-x_1)^2\\\\
    1
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    1
    \end{bmatrix}
$$


$$\frac{\partial g_2}{\partial x}=
    \begin{bmatrix}
    0\\\\
    -1
    \end{bmatrix}
$$

These are linearly dependent and this is the reason why the KKT conditions are not satisfied at the optimal solution.

### Problem 3 (30 Points)

Find a local solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2,x_3} & \quad  f=x_1x_2+x_2x_3+x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$ 

Use two methods: reduced gradient and Lagrange multipliers.

---
#### Solution

First, we will convert the problem to a minimization problem. Therefore, we have:

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=-x_1x_2-x_2x_3-x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$


#### a) Reduced Gradient Method


Number of variables: $n=3$

Number of constraints: $m=1$

Degrees of freedom: $dof=n-m=2$

<br>
We choose the following decision and state variables:
<br>
Decision variables: $d= \begin{bmatrix}
    x_2 &
    x_3
    \end{bmatrix}^T
$
<br>
State variables: $s=[x_1]$

We have that:

$$\frac{\partial f}{\partial d}=
    \begin{bmatrix}
    \frac{\partial f}{\partial x_2}\\\\
    \frac{\partial f}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 - x_3\\\\
    -x_2 - x_1
    \end{bmatrix}
$$


$$
\frac{\partial f}{\partial s}=\frac{\partial f}{\partial x_1}=-x_2 - x_3
$$


$$
\frac{\partial h}{\partial s}=\frac{\partial f}{\partial x_1}=1
$$


$$\frac{\partial h}{\partial d}=
    \begin{bmatrix}
    \frac{\partial h}{\partial x_2}\\\\
    \frac{\partial h}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    1\\\\
    1
    \end{bmatrix}
$$


According to the Reduced Gradient Method we have that:

$$\begin{equation*}
    \frac{df}{dd} = \frac{\partial f}{\partial d}-\frac{\partial f}{\partial s}\cdot(\frac{\partial h}{\partial s})^{-1}\cdot \frac{\partial h}{\partial d}
  \end{equation*} 
$$


$$\Rightarrow\frac{df}{dd}=
    \begin{bmatrix}
    \frac{\partial h}{\partial x_2}\\\\
    \frac{\partial h}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 - x_3\\\\
    -x_2 - x_1
    \end{bmatrix} - (-x_2 - x_3)
    \begin{bmatrix}
    1\\\\
    1
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 + x_2\\\\
    -x_1 + x_3
    \end{bmatrix}
$$


If we set

$$
\frac{df}{dd}=0
$$


$$
\Rightarrow x_1=x_2 \text{ and }  x_1=x_3
$$


From the constraint $h=0$, we get that the stationary point is: 

$$x_1=x_2=x_3=1$$

We also need to check the 2nd order conditions in order to determine whether this stationary point is a local minimum.

The 2nd order conditions are:

$$\frac{d^2h}{dd^2}=
    \begin{bmatrix}
    I & \frac{ds}{dd}^T
    \end{bmatrix}
    \begin{bmatrix}
    \frac{\partial^2 h}{\partial d^2} & \frac{\partial^2 h}{\partial d\partial s}\\\\
    \frac{\partial^2 h}{\partial s\partial d} & \frac{\partial^2 h}{\partial s^2}
    \end{bmatrix}
    \begin{bmatrix}
    I \\\\
    \frac{ds}{dd}
    \end{bmatrix}+ \frac{\partial h}{\partial s}\frac{d^2 s}{dd^2}=0
$$



$$\frac{d^2h}{dd^2}=
    \begin{bmatrix}
    I & \frac{ds}{dd}^T
    \end{bmatrix}
    \begin{bmatrix}
    \frac{\partial^2 f}{\partial d^2} & \frac{\partial^2 f}{\partial d\partial s}\\\\
    \frac{\partial^2 f}{\partial s\partial d} & \frac{\partial^2 f}{\partial s^2}
    \end{bmatrix}
    \begin{bmatrix}
    I \\\\
    \frac{ds}{dd}
    \end{bmatrix}+ \frac{\partial f}{\partial s}\frac{d^2 s}{dd^2}
$$


The derivatives for this problem are:

$$\frac{ds}{dd}=
    -(\frac{\partial h}{\partial s})^{-1}\cdot\frac{\partial h}{\partial d}=
    \begin{bmatrix}
    -1\\\\
    -1
    \end{bmatrix}
$$


$$\frac{\partial^2 h}{\partial d^2}=
    \begin{bmatrix}
    \frac{\partial^2 h}{\partial x_2^2} & \frac{\partial^2 h}{\partial x_2\partial x_3}\\\\
    \frac{\partial^2 h}{\partial x_3\partial x_2} & \frac{\partial^2 h}{\partial x_3^2}
    \end{bmatrix}=
    \begin{bmatrix}
    0 & 0\\\\
    0 & 0
    \end{bmatrix}
$$


$$
\frac{\partial^2 h}{\partial s^2}=0
$$


$$\frac{\partial^2 h}{\partial d\partial s}=\frac{\partial^2 h}{\partial s\partial d}=
    \begin{bmatrix}
    0 \\\\
    0 
    \end{bmatrix}
$$





$$\frac{\partial^2 f}{\partial d^2}=
    \begin{bmatrix}
    \frac{\partial^2 f}{\partial x_2^2} & \frac{\partial^2 f}{\partial x_2\partial x_3}\\\\
    \frac{\partial^2 f}{\partial x_3\partial x_2} & \frac{\partial^2 f}{\partial x_3^2}
    \end{bmatrix}=
    \begin{bmatrix}
    0 & -1\\\\
    -1 & 0
    \end{bmatrix}
$$


$$
\frac{\partial^2 f}{\partial s^2}=0
$$


$$\frac{\partial^2 f}{\partial d\partial s}=\frac{\partial^2 f}{\partial s\partial d}=
    \begin{bmatrix}
    -1 \\\\
    -1 
    \end{bmatrix}
$$


So,

$$\frac{d^2 h}{dd^2}=
    \begin{bmatrix}
    1 & -1 & -1
    \end{bmatrix}
    \begin{bmatrix}
    0 & 0 & 0\\\\
    0 & 0 & 0\\\\
    0 & 0 & 0
    \end{bmatrix}
    \begin{bmatrix}
    1\\\\
    -1\\\\
    -1
    \end{bmatrix}+\frac{d^2 s}{dd^2}=0
$$


$$
\Rightarrow \frac{d^2 s}{dd^2}=0
$$

and

$$\frac{d^2 f}{dd^2}=
    \begin{bmatrix}
    1 & -1 & -1
    \end{bmatrix}
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 & 0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}
    \begin{bmatrix}
    1\\\\
    -1\\\\
    -1
    \end{bmatrix}+0=2>0
$$

Therefore, the stationary point is a local solution (local minimum). Hence, it maximizes the original problem.



#### b) Lagrangian Method

The Lagrangian is:

$$
\mathcal{L}=-x_1x_2 - x_2x_3 - x_1x_3 +\lambda(x_1+x_2+x_3-3) 
$$


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}\\\\
    \frac{\partial \mathcal{L}}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_2-x_3+\lambda \\\\
    -x_1-x_3+\lambda \\\\
    -x_2-x_1+\lambda 
    \end{bmatrix}=
    \begin{bmatrix}
    0 \\\\
    0 \\\\
    0 
    \end{bmatrix}
$$



$$
\frac{\partial \mathcal{L}}{\partial \lambda}=x_1+x_2+x_3-3
$$


Therefore, the stationary point is:
$$
 x_1=x_2=x_3=1 \text{ and } \lambda=2
$$

We need to also check the 2nd-order condition.

$$L_{xx}=\frac{\partial^2 \mathcal{L}}{\partial x^2}=
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 &0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}
$$


$$\frac{\partial h}{\partial x}=
    \begin{bmatrix}
    1\\\\
    1\\\\
    1
    \end{bmatrix}
$$


The feasible perturbations are given by:


$$
(\frac{\partial h}{\partial x})^Tdx=0
$$


$$
\Rightarrow
    \begin{bmatrix}
    1 & 1 & 1
    \end{bmatrix}\cdot
    \begin{bmatrix}
    dx_1\\\\
    dx_2\\\\
    dx_3
    \end{bmatrix}=0
$$


$$
\Rightarrow dx_1= -dx_2-dx_3
$$


Therefore, the 2nd-order condition is:


$$dx^T\cdot L_{xx}\cdot dx=
    \begin{bmatrix}
    dx_1 & dx_2 & dx_3
    \end{bmatrix}\cdot
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 &0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}\cdot
    \begin{bmatrix}
    dx_1\\\\
    dx_2\\\\
    dx_3
    \end{bmatrix}
$$


$$\Rightarrow dx^T\cdot L_{xx}\cdot dx=
    (dx_2+dx_3)^2+ dx_2^2 + dx_3^2 \ge 0
$$

However, since it must be $dx\neq0$, all three components cannot be zero simultaneously. Hence,


$$dx^T\cdot L_{xx}\cdot dx>0 
$$


Therefore, the stationary point is a local solution.

### Problem 4 (20 Points)

Use reduced gradient to	find the value(s) of the parameter $b$ for which the point $x_1=1$, $x_2=2$ is the solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2} & \quad  f=2x_{1} + bx_2\\
{\text{subject to }} & \quad g_1 = x_{1}^{2}+ x_{2}^{2}-5\leq 0\\
& \quad g_2= x_1- x_2-2\leq 0.
\end{aligned}
$$ 

<br>

---

### Solution

We first convert the given problem to a minimization one:


$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-2x_1 - bx_2\\
{\text{subject to }} & \quad g_1 = x_{1}^{2}+ x_{2}^{2}-5\leq 0\\
& \quad g_2= x_1- x_2-2\leq 0.
\end{aligned}
$$ 


At the solution $(x_1,x_2)=(1,2)$, the constraints are:

$$
g_1 = x_{1}^{2}+ x_{2}^{2}-5=1+4-5=0\\
g_2= x_1- x_2-2=1-2-2=-3 < 0
$$ 

So, $g_1$ is an active constraint and $g_2$ is non-active. Therefore, at the given solution, the problem can be stated as:


$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-2x_1 - bx_2\\
{\text{subject to }} & \quad h_1 = x_{1}^{2}+ x_{2}^{2}-5= 0
\end{aligned}
$$


We have that $n=2, m=1, dof=1$ and we choose the following decision and state variables:

We choose the following decision and state variables:
<br>
Decision variable: $d = x_1
$
<br>
State variable: $s=x_2$

We have that:

$$
\frac{\partial f}{\partial d}=\frac{\partial f}{\partial x_1}=-2
$$


$$
\frac{\partial f}{\partial s}=\frac{\partial f}{\partial x_2}=-b
$$


$$
\frac{\partial h}{\partial s}=\frac{\partial h}{\partial x_2}=2x_2
$$


$$
\frac{\partial h}{\partial d}=\frac{\partial h}{\partial x_1}=2x_1
$$


So, according to the Reduced Gradient Method we have that:

$$\begin{equation*}
    \frac{df}{dd} = \frac{\partial f}{\partial d}-\frac{\partial f}{\partial s}\cdot(\frac{\partial h}{\partial s})^{-1}\cdot \frac{\partial h}{\partial d}
  \end{equation*} 
$$


$$
\Rightarrow \frac{df}{dd}=-2-(-b)\frac{1}{2x_2}2x_1=-2+b\frac{x_1}{x_2}
$$


If we set the gradient to zero, we get that:

$$
\frac{df}{dd}=0\\
