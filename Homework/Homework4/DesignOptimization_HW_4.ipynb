### Problem 1 (10 Points)

Sketch graphically the problem 

$$
\begin{aligned}
\min_{x_1,x_2} & \quad f({\bf x})=(x_1+1)^2+(x_2-2)^2\\
{\text{subject to }} & \quad g_1 = x_1-2\leq 0,{\quad} g_3 = -x_1\leq 0,\\
& \quad g_2 = x_2-1\leq 0, {\quad} g_4 = -x_2\leq 0.
\end{aligned}
$$

Find the optimum graphically. Determine directions of feasible descent at the corner points of the feasible domain. Show the gradient directions of $f$ and $g_i$s at these points. Verify graphical results analytically using the KKT conditions.

---
#### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle



def f(x):
    """The objective function."""
    return (x[0]+1)**2 + (x[1]-2)**2

def plot_contour(f):
    """Plots objective function contour, constraints and gradient directions."""

    # Create mesh
    x1 = np.linspace(-4,4,100)
    x2 = np.linspace(-4,4,100)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f([X1, X2])

    # Contour
    fig, ax = plt.subplots(figsize = (12,8))
    plt.contourf(X1,X2,Z,10, cmap='OrRd')
    ax.set(xlabel='x1', ylabel='x2')
    ax.set_title('Contour of f(x), Constraints and Gradient Directions')
    plt.colorbar()
    # Axes
    ax.axhline(y=0, linewidth=1, color='k')  
    ax.axvline(x=0, linewidth=1, color='k')
    # Constraints
    ax.hlines(y=1, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.hlines(y=0, xmin=0, xmax=2, linewidth=1.5, color='k')
    ax.vlines(x=0, ymin=0, ymax=1, linewidth=1.5, color='k')
    ax.vlines(x=2, ymin=0, ymax=1, linewidth=1.5, color='k')
    # Optimum
    ax.plot(0, 1, 'o', color='r', label='Constrained Optimum')
    ax.plot(-1, 2, 'o', color='#ffb3b0', label='Obj. Funtion Optimum')
    # Gradient directions
    c = 5  # Factor to reduce the vector magnitudes
    plt.arrow(0,1, 2/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,1)
    plt.arrow(0,1, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 2/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (0,0)
    plt.arrow(0,0, -1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(0,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 6/c, -4/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,0)
    plt.arrow(2,0, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,0, 0, -1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 6/c, -2/c, head_width=0.1, head_length=0.2, fc='#ab4343', color='#ab4343')  # At point (2,1)
    plt.arrow(2,1, 1/c, 0, head_width=0.1, head_length=0.2,fc='k')
    plt.arrow(2,1, 0, 1/c, head_width=0.1, head_length=0.2,fc='k')
    plt.legend()
    # Anotations and feasible domain fill
    ax.annotate(r'$\frac{\partial f}{\partial x}$', xy=(2.8, 1),color='#ab4343', size='x-large')
    ax.annotate('feasible domain', xy=(0.2, 0.2),color='#702c2c', size='large')
    ax.add_patch(Rectangle((0, 0), 2, 1, color="#702c2c", alpha=0.2))
    plt.show()


plot_contour(f)
```

#### KKT Conditions

The Lagrangian is defined as:

$$\begin{aligned}
    \mathcal{L}=(x_1+1)^2+(x_2-2)^2+\mu_1 (x_1-2) +\mu_2 (x_2-1) +\mu_3 (-x_1) +\mu_4 (-x_2)
  \end{aligned}
$$

The gradient of the Lagrangian is:


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}
    \end{bmatrix}=
    \begin{bmatrix}
    2(x_1+1) + \mu_1 - \mu_3\\\\
    2(x_2-2) + \mu_2 - \mu_4
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    0
    \end{bmatrix}
$$


Since there are 4 constraints, there are $2^4$ possibilities for each $\mu_i >0$ or $\mu_i =0$. However, since we have solved the problem graphically, we will only check this point to see whether it satisfies the KKT conditions.

At the point $(x_1,x_2)=(0,1)$, we have the following active constraints: $g_2$ and $g_3$. Therefore, 

$$\begin{aligned}
    \mu_2 >0, \mu_3 >0, \mu_1 =\mu_4 =0
  \end{aligned}
$$


<br>
$$\Rightarrow 
  \begin{equation*}
  \left\{\begin{aligned}
    2(x_1+1) - \mu_3 =0\\
    2(x_2-2) + \mu_2 =0\\
    -x_1 = 0\\
    x_2 -1 = 0
  \end{aligned}
  \right.\end{equation*} 
$$


<br>
$$\Rightarrow 
  \begin{equation*}
    \mu_2 = 2, \mu_3 = 2 >0
  \end{equation*} 
$$

This point satisfies the KKT conditions, hence it is a local solution.

The feasible domain (rectangle) and the objective function (circular contours) are convex, thus the problem is convex. Therefore, since the problem is convex, the local solution $(x_1,x_2)=(0,1)$ is the global optimum.


### Problem 2 (10 Points)

Graph the problem 

$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-x_1\\
{\text{subject to }} & \quad g_1=x_2-(1-x_1)^3\leq 0{\quad} {\rm and}{\quad} x_2\geq 0.
\end{aligned}
$$ 

Find the solution graphically. Then apply the optimality conditions. Can you find a solution based on the optimality conditions? Why? (From Kuhn and Tucker, 1951.)

---
#### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle



def f(x):
    """The objective function."""
    return -x[0]

def plot_contour(f):
    """Plots objective function contour, constraints and gradient directions."""

    # Create mesh
    x1 = np.linspace(-4,4,100)
    x2 = np.linspace(-4,4,100)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f([X1, X2])

    # Contour
    fig, ax = plt.subplots(figsize = (12,8))
    plt.contourf(X1,X2,Z,10, cmap='OrRd')
    ax.set(xlabel='x1', ylabel='x2')
    ax.set_title('Contour of f(x) and Constraints')
    plt.colorbar()
    # Axes
    ax.axhline(y=0, linewidth=1, color='k')  
    ax.axvline(x=0, linewidth=1, color='k')
    # Constraints
    y = (1-x1)**3
    ax.plot(x1, y, color='k')
    ax.hlines(y=0, xmin=-4, xmax=4, linewidth=2, color='k')
    # Optimum
    ax.plot(1, 0, 'o', color='#702c2c', label='Constrained Optimum')    
    # Anotations and feasible domain fill
    ax.annotate('feasible domain', xy=(-3, 1),color='k', size='large')
    ax.fill_between(x1, y, where=x1<1, color='#539ecd', alpha=0.3)
    ax.set_ylim([-4, 4])
    plt.legend()
    plt.show()


plot_contour(f)
```

From the graph above, we observe that the solution is at $(x_1,x_2)=(1,0)$.

The Lagrangian is defined as:

$$\begin{aligned}
    \mathcal{L}= -x_1+\mu_1 (x_2-(1-x_1)^3) +\mu_2 (-x_2) 
  \end{aligned}
$$

The gradient of the Lagrangian is:


$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}
    \end{bmatrix}=
    \begin{bmatrix}
    -1+3 \mu_1 (1-x_1)^2\\\\
    \mu_1 - \mu_2
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    0
    \end{bmatrix}
$$


At $(x_1,x_2)=(1,0)$, which we know is the solution, both constraints are active, thus:

$$\begin{aligned}
    \mu_1 >0, \mu_2 >0
  \end{aligned}
$$


Therefore, from the gradient of the Lagrangian we get that:

<br>
$$\Rightarrow 
  \begin{equation*}
  \left\{\begin{aligned}
    -1  =0\\
    \mu_1 =\mu_2
  \end{aligned}
  \right.\end{equation*} 
$$

This does not hold, hence this point does not satisfy the KKT conditions. We observe that an optimal solution does exist but the KKT conditions are not satisfied.

The constraint gradients are computed as follows:

<br>
$$\frac{\partial g_1}{\partial x}=
    \begin{bmatrix}
    3(1-x_1)^2\\\\
    1
    \end{bmatrix}=
    \begin{bmatrix}
    0\\\\
    1
    \end{bmatrix}
$$


<br>
$$\frac{\partial g_2}{\partial x}=
    \begin{bmatrix}
    0\\\\
    -1
    \end{bmatrix}
$$

These are linearly dependent and this is the reason why the KKT conditions are not satisfied at the optimal solution.

### Problem 3 (30 Points)

Find a local solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2,x_3} & \quad  f=x_1x_2+x_2x_3+x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$ 

Use two methods: reduced gradient and Lagrange multipliers.

---
#### Solution

First, we will convert the problem to a minimization problem. Therefore, we have:

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=-x_1x_2-x_2x_3-x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$


#### a) Reduced Gradient Method


Number of variables: $n=3$

Number of constraints: $m=1$

Degrees of freedom: $dof=n-m=2$

<br>
We choose the following decision and state variables:
<br>
Decision variables: $d= \begin{bmatrix}
    x_2 &
    x_3
    \end{bmatrix}^T
$
<br>
State variables: $s=x_1$

We have that:

<br>
$$\frac{\partial f}{\partial d}=
    \begin{bmatrix}
    \frac{\partial f}{\partial x_2}\\\\
    \frac{\partial f}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 - x_3\\\\
    -x_2 - x_1
    \end{bmatrix}
$$


<br>
$$
\frac{\partial f}{\partial s}=\frac{\partial f}{\partial x_1}=-x_2 - x_3
$$

<br>
$$
\frac{\partial h}{\partial s}=\frac{\partial f}{\partial x_1}=1
$$

<br>
$$\frac{\partial h}{\partial d}=
    \begin{bmatrix}
    \frac{\partial h}{\partial x_2}\\\\
    \frac{\partial h}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    1\\\\
    1
    \end{bmatrix}
$$


According to the Reduced Gradient Method we have that:
<br>
$$\begin{equation*}
    \frac{df}{dd} = \frac{\partial f}{\partial d}-\frac{\partial f}{\partial s}\cdot(\frac{\partial h}{\partial s})^{-1}\cdot \frac{\partial h}{\partial d}
  \end{equation*} 
$$

<br>
$$\Rightarrow\frac{df}{dd}=
    \begin{bmatrix}
    \frac{\partial h}{\partial x_2}\\\\
    \frac{\partial h}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 - x_3\\\\
    -x_2 - x_1
    \end{bmatrix} - (-x_2 - x_3)
    \begin{bmatrix}
    1\\\\
    1
    \end{bmatrix}=
    \begin{bmatrix}
    -x_1 + x_2\\\\
    -x_1 + x_3
    \end{bmatrix}
$$


If we set
<br>
$$
\frac{df}{dd}=0
$$

<br>
$$
\Rightarrow x_1=x_2 \text{ and }  x_1=x_3
$$


From the constraint $h=0$, we get that the stationary point is: 

$$x_1=x_2=x_3=1$$

We also need to check the 2nd order conditions in order to determine whether this stationary point is a local minimum.

The 2nd order conditions are:

$$\frac{d^2h}{dd^2}=
    \begin{bmatrix}
    I & \frac{ds}{dd}^T
    \end{bmatrix}
    \begin{bmatrix}
    \frac{\partial^2 h}{\partial d^2} & \frac{\partial^2 h}{\partial d\partial s}\\\\
    \frac{\partial^2 h}{\partial s\partial d} & \frac{\partial^2 h}{\partial s^2}
    \end{bmatrix}
    \begin{bmatrix}
    I \\\\
    \frac{ds}{dd}
    \end{bmatrix}+ \frac{\partial h}{\partial s}\frac{d^2 s}{dd^2}=0
$$


<br>
$$\frac{d^2h}{dd^2}=
    \begin{bmatrix}
    I & \frac{ds}{dd}^T
    \end{bmatrix}
    \begin{bmatrix}
    \frac{\partial^2 f}{\partial d^2} & \frac{\partial^2 f}{\partial d\partial s}\\\\
    \frac{\partial^2 f}{\partial s\partial d} & \frac{\partial^2 f}{\partial s^2}
    \end{bmatrix}
    \begin{bmatrix}
    I \\\\
    \frac{ds}{dd}
    \end{bmatrix}+ \frac{\partial f}{\partial s}\frac{d^2 s}{dd^2}
$$


The derivatives for this problem are:

$$\frac{ds}{dd}=
    -(\frac{\partial h}{\partial s})^{-1}\cdot\frac{\partial h}{\partial d}=
    \begin{bmatrix}
    -1\\\\
    -1
    \end{bmatrix}
$$

<br>
$$\frac{\partial^2 h}{\partial d^2}=
    \begin{bmatrix}
    \frac{\partial^2 h}{\partial x_2^2} & \frac{\partial^2 h}{\partial x_2\partial x_3}\\\\
    \frac{\partial^2 h}{\partial x_3\partial x_2} & \frac{\partial^2 h}{\partial x_3^2}
    \end{bmatrix}=
    \begin{bmatrix}
    0 & 0\\\\
    0 & 0
    \end{bmatrix}
$$

<br>
$$
\frac{\partial^2 h}{\partial s^2}=0
$$

<br>
$$\frac{\partial^2 h}{\partial d\partial s}=\frac{\partial^2 h}{\partial s\partial d}=
    \begin{bmatrix}
    0 \\\\
    0 
    \end{bmatrix}
$$




<br>
$$\frac{\partial^2 f}{\partial d^2}=
    \begin{bmatrix}
    \frac{\partial^2 f}{\partial x_2^2} & \frac{\partial^2 f}{\partial x_2\partial x_3}\\\\
    \frac{\partial^2 f}{\partial x_3\partial x_2} & \frac{\partial^2 f}{\partial x_3^2}
    \end{bmatrix}=
    \begin{bmatrix}
    0 & -1\\\\
    -1 & 0
    \end{bmatrix}
$$

<br>
$$
\frac{\partial^2 f}{\partial s^2}=0
$$

<br>
$$\frac{\partial^2 f}{\partial d\partial s}=\frac{\partial^2 f}{\partial s\partial d}=
    \begin{bmatrix}
    -1 \\\\
    -1 
    \end{bmatrix}
$$


So,

$$\frac{d^2 h}{dd^2}=
    \begin{bmatrix}
    1 & -1 & -1
    \end{bmatrix}
    \begin{bmatrix}
    0 & 0 & 0\\\\
    0 & 0 & 0\\\\
    0 & 0 & 0
    \end{bmatrix}
    \begin{bmatrix}
    1\\\\
    -1\\\\
    -1
    \end{bmatrix}+\frac{d^2 s}{dd^2}=0
$$

<br>
$$
\Rightarrow \frac{d^2 s}{dd^2}=0
$$

and

$$\frac{d^2 f}{dd^2}=
    \begin{bmatrix}
    1 & -1 & -1
    \end{bmatrix}
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 & 0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}
    \begin{bmatrix}
    1\\\\
    -1\\\\
    -1
    \end{bmatrix}+0=2>0
$$

Therefore, the stationary point is a local solution (local minimum). Hence, it maximizes the original problem.



#### b) Lagrangian Method

The Lagrangian is:

<br>
$$
\mathcal{L}=-x_1x_2 - x_2x_3 - x_1x_3 +\lambda(x_1+x_2+x_3-3) 
$$

<br>
$$\frac{\partial \mathcal{L}}{\partial x}=
    \begin{bmatrix}
    \frac{\partial \mathcal{L}}{\partial x_1}\\\\
    \frac{\partial \mathcal{L}}{\partial x_2}\\\\
    \frac{\partial \mathcal{L}}{\partial x_3}
    \end{bmatrix}=
    \begin{bmatrix}
    -x_2-x_3+\lambda \\\\
    -x_1-x_3+\lambda \\\\
    -x_2-x_1+\lambda 
    \end{bmatrix}=
    \begin{bmatrix}
    0 \\\\
    0 \\\\
    0 
    \end{bmatrix}
$$


<br>
$$
\frac{\partial \mathcal{L}}{\partial \lambda}=x_1+x_2+x_3-3
$$


Therefore, the stationary point is:
$$
 x_1=x_2=x_3=1 \text{ and } \lambda=2
$$

We need to also check the 2nd-order condition.

$$L_{xx}=\frac{\partial^2 \mathcal{L}}{\partial x^2}=
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 &0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}
$$

<br>
$$\frac{\partial h}{\partial x}=
    \begin{bmatrix}
    1\\\\
    1\\\\
    1
    \end{bmatrix}
$$


The feasible perturbations are given by:

<br>
$$
(\frac{\partial h}{\partial x})^Tdx=0
$$

<br>
$$
\Rightarrow
    \begin{bmatrix}
    1 & 1 & 1
    \end{bmatrix}\cdot
    \begin{bmatrix}
    dx_1\\\\
    dx_2\\\\
    dx_3
    \end{bmatrix}=0
$$

<br>
$$
\Rightarrow dx_1= -dx_2-dx_3
$$


Therefore, the 2nd-order condition is:

<br>
$$dx^T\cdot L_{xx}\cdot dx=
    \begin{bmatrix}
    dx_1 & dx_2 & dx_3
    \end{bmatrix}\cdot
    \begin{bmatrix}
    0 & -1 & -1\\\\
    -1 &0 & -1\\\\
    -1 & -1 & 0
    \end{bmatrix}\cdot
    \begin{bmatrix}
    dx_1\\\\
    dx_2\\\\
    dx_3
    \end{bmatrix}
$$

<br>
$$\Rightarrow dx^T\cdot L_{xx}\cdot dx=
    (dx_2+dx_3)^2+ dx_2^2 + dx_3^2 \ge 0
$$

However, since it must be $dx\neq0$, all three components cannot be zero simultaneously. Hence,


$$dx^T\cdot L_{xx}\cdot dx>0 
$$


Therefore, the stationary point is a local solution.

### Problem 4 (20 Points)

Use reduced gradient to	find the value(s) of the parameter $b$ for which the point $x_1=1$, $x_2=2$ is the solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2} & \quad  f=2x_{1} + bx_2\\
{\text{subject to }} & \quad g_1 = x_{1}^{2}+ x_{2}^{2}-5\leq 0\\
& \quad g_2= x_1- x_2-2\leq 0.
\end{aligned}
$$ 

<br>

---

### Solution

We first convert the given problem to a minimization one:


$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-2x_1 - bx_2\\
{\text{subject to }} & \quad g_1 = x_{1}^{2}+ x_{2}^{2}-5\leq 0\\
& \quad g_2= x_1- x_2-2\leq 0.
\end{aligned}
$$ 


At the solution $(x_1,x_2)=(1,2)$, the constraints are:

$$
g_1 = x_{1}^{2}+ x_{2}^{2}-5=1+4-5=0\\\\
g_2= x_1- x_2-2=1-2-2=-3 < 0
$$ 

So, $g_1$ is an active constraint and $g_2$ is non-active. Therefore, at the given solution, the problem can be stated as:


$$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-2x_1 - bx_2\\
{\text{subject to }} & \quad h_1 = x_{1}^{2}+ x_{2}^{2}-5= 0
\end{aligned}
$$


We have that $n=2, m=1, dof=1$ and we choose the following decision and state variables:

Decision variable: $d = x_1
$
<br>
State variable: $s=x_2$

We have that:

<br>
$$
\frac{\partial f}{\partial d}=\frac{\partial f}{\partial x_1}=-2
$$

<br>
$$
\frac{\partial f}{\partial s}=\frac{\partial f}{\partial x_2}=-b
$$

<br>
$$
\frac{\partial h}{\partial s}=\frac{\partial h}{\partial x_2}=2x_2
$$

<br>
$$
\frac{\partial h}{\partial d}=\frac{\partial h}{\partial x_1}=2x_1
$$


So, according to the Reduced Gradient Method we have that:

$$\begin{equation*}
    \frac{df}{dd} = \frac{\partial f}{\partial d}-\frac{\partial f}{\partial s}\cdot(\frac{\partial h}{\partial s})^{-1}\cdot \frac{\partial h}{\partial d}
  \end{equation*} 
$$

<br>
$$
\Rightarrow \frac{df}{dd}=-2-(-b)\frac{1}{2x_2}2x_1=-2+b\frac{x_1}{x_2}
$$


If we set the gradient to zero, we get that:

$$
\frac{df}{dd}=0\\
\Rightarrow b=4
$$


Therefore, at the solution the value of the parameter $b$ is $b=4$.


### Problem 5 (30 Points)

Find the solution for 

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=x_{1}^{2}+x_{2}^{2}+x_{3}^{2}\\
{\text{subject to }} & \quad h_1 = x_{1}^{2}/4+x_{2}^{2}/5+x_{3}^{2}/25-1=0\\
& \quad h_2 = x_1+x_2-x_3= 0,
\end{aligned}
$$ 

by implementing the generalized reduced gradient algorithm.

```python
"""Solves non-linear optimization problems with the Generalized Reduced Gradient method."""

import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import seaborn as sns


def obj_fun():
    """The objective function to be optimized.

    Returns:
      - obj(sympy.core.add.Add): The objective function in symbolic form
      - variables(tuple):        The variables
    """
    x1, x2, x3 = symbols('x1 x2 x3')
    variables = (x1, x2, x3)
    obj = x1**2 + x2**2 + x3**2
    return obj, variables


def constraints():
    """The problem equality constraints h(x)=0.

    Returns:
      - h(np.ndarray): The equality constraints
    """
    x1, x2, x3 = symbols('x1 x2 x3')
    h = np.array([x1**2/4 + x2**2/5 + x3**2/25 -1,
                  x1 + x2 - x3])
    return h


def decision_and_state_vars(variables, h):
    """Separates the variables into decision and state variables.

    The first n-m vars are chosen as decision vars and the rest as state vars.
    Inputs:
      - variables(tuple):      The variables
      - h(np.ndarray):         The equality constraints
    Returns:
      - dec_vars(tuple):       The chosen decision variables
      - state_vars(tuple):     The chosen state variables
    """

    n = len(variables)  # Number of variables
    m = len(h)          # Number of constraints
    dof = n-m           # Number of decision variables
    dec_vars = tuple([var for var in variables[0:dof]])
    state_vars = tuple([var for var in variables[dof:]])
    return dec_vars, state_vars


def df_dd_sym(f, dec_vars):
    """Computes the partial derivatives of f wrt the decision vars d.

    Inputs:
      - f(sympy.core.add.Add): The objective function in symbolic form
      - dec_vars(tuple):       The decision variables
    """
    return np.array([diff(f, var) for var in dec_vars])


def df_ds_sym(f, state_vars):
    """Computes the partial derivatives of f wrt the state vars s.

    Inputs:
      - f(sympy.core.add.Add): The objective function in symbolic form
      - state_vars(tuple):     The state variables
    """
    return np.array([diff(f, var) for var in state_vars])


def dh_dd_sym(h, dec_vars):
    """Computes the partial derivatives of h wrt the decision vars d.

    Inputs:
      - h(np.ndarray):   The equality constraints
      - dec_vars(tuple): The decision variables
    """
    partialh_partiald = np.zeros([len(h), len(dec_vars)], dtype=object)
    for i in range(len(h)):
        partialh_partiald[i,:] = np.array([diff(h[i], var) for var in dec_vars]).reshape((1, len(dec_vars)))  # dhi_dd
    return partialh_partiald


def dh_ds_sym(h, state_vars):
    """Computes the partial derivatives of h wrt the state vars s.

    Inputs:
      - h(np.ndarray):     The equality constraints
      - state_vars(tuple): The state variables
    """
    partialh_partials = np.zeros([len(h), len(state_vars)], dtype=object)
    for i in range(len(h)):
        # dhi_ds
        partialh_partials[i, :] = np.array([diff(h[i], var) for var in state_vars]).reshape((1, len(state_vars)))
    return partialh_partials


def newton_raphson(h_k, dh_ds_k, d_k1_j, s_k1_j):
    """Compute s_k1 such that it satisfies h=0 via the Newton-Raphson method.

    Inputs:
      - h_k(function):      Equality constraints
      - dh_ds_k(function):  Partial derivatives of h wrt s
      - d_k1_j(np.ndarray): The decision variables
      - s_k1_j(np.ndarray): The state variables
    Returns:
      - s_k1_j(np.ndarray): The state variables that satisfies h=0 after we move to d_k1
    """

    max_iter = 1000                                    # Maximum iterations before halting
    eps = 1e-6                                         # Tolerance (termination criterion)

    x_k1_j = np.concatenate((d_k1_j, s_k1_j), axis=0)  # Current point
    h_current = h_k(x_k1_j)                            # Current constraint values
    j = 0                                              # Current iteration

    while np.linalg.norm(h_current) > eps and j < max_iter:
        # Compute new guess
        s_k1_j1 = s_k1_j - np.dot(np.linalg.inv(dh_ds_k(s_k1_j)), h_current)

        # Update values
        s_k1_j = s_k1_j1
        x_k1_j = np.concatenate((d_k1_j, s_k1_j), axis=0)
        h_current = h_k(x_k1_j)
        j += 1

    if np.linalg.norm(h_current) > eps:
        print('Newton-Raphson algorithm to find s_k1 did not converge!')
        print('Constraint magnitude: {}'.format(np.linalg.norm(h_current)))

    return s_k1_j


def line_search(f_k, dfdd_k, ds_k, d_k, s_k):
    """Line search to compute best step size a.

    Inputs:
      - f_k(function):    Objective function
      - dfdd_k(function): Reduced gradient
      - ds_k(function):   Change in s
      - d_k(np.ndarray):  The decision variables at current point
      - s_k(np.ndarray):  The state variables at current point
    Returns:
      - a(float): The best step size along the search direction
    """

    # Initialize parameters
    a = 1.0                                     # Step size
    b = 0.5                                     # Reduction factor
    t = 0.5                                     # Relaxation factor

    # Compute values at current point
    x_k = np.concatenate((d_k, s_k), axis=0)    # Current point
    f_xk = f_k(x_k)                             # Function value at current point
    dfdd_xk = np.array(dfdd_k(x_k))             # Reduced gradient at current point
    dfdd_xk_norm2 = np.linalg.norm(dfdd_xk)**2  # Reduced gradient norm^2 at current point
    ds_xk = np.array(ds_k(x_k))                 # ds at current point

    # Compute function value for current a
    d = d_k - a*dfdd_xk
    s = s_k + a*ds_xk
    x_k_a = np.concatenate((d, s), axis=0)
    fa = f_k(x_k_a)

    # 1st-order Taylor series approx. at current point as a function of a
    phi = lambda a: f_xk - a*t*dfdd_xk_norm2

    while fa > phi(a) and a > 1e-4:
        a = a*b  # Reduce a (Backtrack)

        # Compute function value for current a
        x_k_a = np.concatenate((d_k - a*dfdd_xk, s_k + a*ds_xk), axis=0)
        fa = f_k(x_k_a)

    return a


def plot_fun_history(fun_history):
    """Plot function values vs iteration.

    Inputs:
      - fun_history(list): Function values at each iteration
    """

    sns.set_theme()
    plt.figure()
    plt.plot(fun_history)
    plt.xlabel('Iteration')
    plt.ylabel('Function values')
    plt.title('Function values vs iteration')
    plt.show()
    return


def plot_error_history(error_history):
    """Plot reduced gradient norm squared values vs iteration.

    Inputs:
      - error_history(list): Reduced gradient norm squared values at each iteration
    """

    sns.set_theme()
    plt.figure()
    plt.plot(error_history)
    plt.xlabel('Iteration')
    plt.ylabel('Reduced gradient norm squared')
    plt.title('Error vs iteration')
    plt.show()
    return


def plot_convergence(fun_history):
    """Plot log-linear convergence.

    Function log errors vs iteration.
    Inputs:
      - fun_history(list): Function values at each iteration
    """
    log_error = np.log10(fun_history-fun_history[-1])

    sns.set_theme()
    plt.figure()
    plt.plot(log_error)
    plt.xlabel('Iteration')
    plt.ylabel('Error $\log(f-f^\star)$')
    plt.title('log-linear convergence plot')
    plt.show()
    print()
    return


def generalized_reduced_gradient(x0, eps, max_iter):
    """Solves non-linear optimization problems with the Generalized Reduced Gradient (GRG) method.

    For minimization problems with equality constraints.
    Inputs:
      - x0(tuple):       Initial variable values
      - eps(float):      Tolerance
      - max_iter(int):   Maximum iterations before halting
    Returns:
      - x_k(np.ndarray): The solution to the minimization problem
    """
    fun_history = []                                   # The function values history
    error_history = []                                 # The reduced gradient norm squared history

    # Create symbolic expressions of the objective, constraints and their derivatives
    fun, vars = obj_fun()                              # Symbolic: Objective function and variables
    h = constraints()                                  # Symbolic: Constraints
    d, s = decision_and_state_vars(vars, h)            # Symbolic: Decision and state variables
    df_dd = df_dd_sym(fun, d)                          # Symbolic: Partial derivatives of f wrt d
    df_ds = df_ds_sym(fun, s)                          # Symbolic: Partial derivatives of f wrt s
    dh_dd = dh_dd_sym(h, d)                            # Symbolic: Partial derivatives of h wrt d
    dh_ds = dh_ds_sym(h, s)                            # Symbolic: Partial derivatives of h wrt s
    dfdd_sym = df_dd - np.linalg.multi_dot([df_ds, Matrix(dh_ds).inv(), dh_dd])  # Symbolic: Reduced gradient
    ds_init_sym = np.transpose(np.linalg.multi_dot([Matrix(dh_ds).inv(), dh_dd, np.transpose(dfdd_sym)]))  # Symbolic: Change in s

    # Create functions from the symbolic expressions
    f_k = lambdify([vars], fun)                        # Objective function
    h_k = lambdify([vars], h)                          # Constraints
    dh_ds_k = lambdify([s], dh_ds)                     # Partial derivatives of h wrt s
    dfdd_k = lambdify([vars], dfdd_sym)                # Reduced gradient
    ds_k = lambdify([vars], ds_init_sym)               # Change in s

    # Initialize variables
    d_k = np.array([v for v in x0[:len(d)]])           # d0
    s_k = np.array([v for v in x0[len(d):len(vars)]])  # s0

    x_k = x0                                           # Current point
    dfdd = np.array(dfdd_k(x_k))                       # Compute reduced gradient
    k = 0                                              # Current iteration

    while np.linalg.norm(dfdd) > eps and k < max_iter:

        fun_history.append(f_k(x_k))                   # Save function value
        error_history.append(np.linalg.norm(dfdd))     # Save reduced gradient norm squared value

        # Step 1: Compute a_k
        a_k = line_search(f_k, dfdd_k, ds_k, d_k, s_k)

        # Step 2: Compute d_k1
        d_k1 = d_k - a_k*dfdd

        # Step 3: Compute initial guess of s_k1
        s_k1_init = s_k + a_k*np.array(ds_k(x_k))

        # Step 4: Compute s_k1
        s_k1 = newton_raphson(h_k, dh_ds_k, d_k1, s_k1_init)

        # Step 5: Update values and Compute reduced gradient dfdd
        d_k = d_k1
        s_k = s_k1
        x_k = np.concatenate((d_k, s_k), axis=0)
        dfdd = np.array(dfdd_k(x_k))
        k += 1

    # Display the optimization problem and the solution
    print('-'*60+'\n\tGeneralized Reduced Gradient (GRG) method\n'+'-'*60)
    print('Objective function: f(x) = {}'.format(fun))
    print('Constraints: h = {}'.format(h))
    print('Decision vars: d = {} \nState vars: s = {}'.format(d, s))
    print('-'*60)
    print('Number of iterations: {}'.format(k-1))
    print('Error (Reduced gradient norm squared): dfdd = {}'.format(dfdd))
    print('Min function value: f(x) = {}'.format(f_k(x_k)))
    print('The solution to the minimization problem is: \nx = {}'.format(x_k))
    print('-'*60 + '\n')

    # Plots
    plot_convergence(fun_history)      # Plot log-linear convergence
    plot_fun_history(fun_history)      # Plot function values vs iteration
    plot_error_history(error_history)  # Plot reduced gradient norm squared vs iteration

    return x_k


if __name__ == "__main__":

    # Parameters
    # --------------------------------------------------------------------------
    epsilon = 1e-3                     # Tolerance (termination criterion)
    max_iterations = 100000            # Maximum iterations before halting
    x0 = (0, np.sqrt(4*24*100)/48, np.sqrt(4*24*100)/48)  # Init. values that satisfy the constraints h=0

    # Solve minimization problem with GRG method
    # --------------------------------------------------------------------------
    x = generalized_reduced_gradient(x0, epsilon, max_iterations)
```
